```python
data_dir = os.path.join(os.getcwd(), "..", "data")
# mega_file = os.path.join(data_dir, "mega_stop_event_1.hdf")
# df = pd.read_hdf(mega_file, "/df", stop=1000)
pickle_path = os.path.join(data_dir, "mega_pickle")
df = pd.read_pickle(pickle_path)
# df = df.head(100000)
# df = pd.read_hdf(mega_file, "/df", stop=1000000)
# df = pd.read_hdf(mega_file, "/df")
```

```python
df = df.drop(columns=["route_number", "time_cat_stop_time", "time_cat_leave_time"])
```

```python
df = df.sort_values(["service_date", "train", "trip_number", "stop_time"])
```

```python
# df = df[~df.isin([np.nan, np.inf, -np.inf]).any(1)]
```

```python
# df = df.reset_index(drop=True)
```

```python
df = df[df["direction"] == 0]
stops = df["location_id"].unique()
print(stops.shape)
```

```python
def df_pop(df, labels):
    popped = df[labels]
    df = df.drop(columns=labels)
    return df, popped


trips = df.groupby(["service_date", "train", "trip_number"])
trips = [trip for _, trip in trips]
x1 = []
x2 = []
y = []
for trip in trips:
    trip = trip.drop_duplicates("location_id")
    trip = trip.drop(
        columns=[
            "service_date",
            "stop_time",
            "leave_time",
            "vehicle_number",
            "location_id",
            "train",
        ]
    )
    trip, date = df_pop(trip, ["day_of_week", "day_of_year"])
    date = date.iloc[0]
    
    
    
    labels = trip["arrival_deviance"].to_list()
   
    trip = pd.get_dummies(trip)

    trip = np.asarray(trip).astype(np.float32)
    date = np.asarray(date).astype(np.float32)
    labels = np.asarray(labels).astype(np.float32)
    
    if trip.shape[0] == 80:
        x1.append(trip)
        x2.append(date)
        y.append(labels)

trips = np.stack(x1)
dates = np.stack(x2)
labels = np.stack(y)

print(trips.shape, date.shape, labels.shape)


trip_shape = (80, 318)
date_shape = (2,)
labels_shape = (80,)
```

```python
# dataset = tf.data.Dataset.from_tensor_slices((trips, labels))

dataset = tf.data.Dataset.from_tensor_slices(
    ({"trip": trips, "date": dates}, labels)
)


batch_size = 1


train_dataset = dataset.shuffle(1000).batch(batch_size)
val_dataset = dataset.skip(1).batch(batch_size)
test_dataset = dataset.skip(10).batch(batch_size)
```

```python
drop_rate = 0.2

trip_input = tf.keras.Input(shape=trip_shape, name="trip")
date_input = tf.keras.Input(shape=date_shape, name="date")

trip = trip_input
date = date_input

# trip = tf.keras.layers.BatchNormalization()(trip)
trip = tf.keras.layers.LSTM(1000, input_shape=trip_shape)(trip)
trip = tf.keras.layers.Flatten()(trip)

date = tf.keras.layers.Embedding(365, 16, input_length=2)(date)
date = tf.keras.layers.Flatten()(date)

pred = tf.keras.layers.concatenate([trip, date])
trip = tf.keras.layers.BatchNormalization()(trip)
pred = tf.keras.layers.Dropout(drop_rate)(pred)
pred = tf.keras.layers.Dense(300, activation="relu")(pred)
pred = tf.keras.layers.Dropout(drop_rate)(pred)
pred = tf.keras.layers.Dense(50, activation="relu")(pred)
pred = tf.keras.layers.Dense(80, name="output")(pred)

model = tf.keras.Model(inputs=[trip_input, date_input], outputs=[pred],)
```

```python
model.compile(
    optimizer="adam",
    loss=tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error"),
    metrics=["mean_absolute_error", "mean_squared_error"],
)
print("total parameters: {:,}".format(model.count_params()))
tf.keras.utils.plot_model(model, "model.png", show_shapes=True, dpi=227)
```

```python
monitor_param = "val_mean_squared_error"


def scheduler(epoch, lr):
    if epoch < 10:
        return lr
    else:
        return lr * tf.math.exp(-0.1)


lr_schedule = tf.keras.callbacks.LearningRateScheduler(scheduler)

nan_terminate = tf.keras.callbacks.TerminateOnNaN()

lr_plateau_reduce = tf.keras.callbacks.ReduceLROnPlateau(
    monitor=monitor_param, patience=3
)

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor=monitor_param, patience=5  # , restore_best_weights=True
)

model_dir = os.path.join(data_dir, "model")
checkpoint_filepath = os.path.join(model_dir, "checkpoint")
checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor=monitor_param,
    mode="min",
    save_best_only=True,
)
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices("GPU")))

model.fit(
    train_dataset,
    epochs=100,
    validation_data=val_dataset,
    callbacks=[early_stopping, checkpoint, nan_terminate],
)
```

```python
model.load_weights(checkpoint_filepath)
```

```python
model.evaluate(test_dataset)
```
